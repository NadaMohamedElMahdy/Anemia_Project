```python
import pandas as pd
```

```python
data=pd.read_csv("Anemia.csv")
data
```

```python
#missing_values
data.isna().sum()
```

```python
data.info()
```

```python
#categorical_data
#(1)-->LAB_TEST            
data.LAB_TEST.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
hot1=OneHotEncoder(sparse=False)
```

```python
hot1_values=hot1.fit_transform(data[["LAB_TEST"]])
```

```python
hot1_values
```

```python
hot1_values_new=pd.DataFrame(hot1_values)
hot1_values_new
```

```python
data.drop("LAB_TEST",axis=1,inplace=True)
```

```python
new_data=pd.concat([data,hot1_values_new],axis=1)
new_data
```

```python
data.REFERENCE_INTERVAL.nunique()
```

```python
#drop_col--->REFERENCE_INTERVAL
new_data.drop("REFERENCE_INTERVAL",axis=1,inplace=True)
new_data
```

```python
#drop_col--->NO
new_data.drop("NO",axis=1,inplace=True)
new_data
```

```python
#(2)--->GENDER
data.GENDER.unique() 
```

```python
new_data.GENDER=data.GENDER.map({'Male':0,'Female':1})
new_data
```

```python
#(3)--->IDENTIFICATION
data.IDENTIFICATION.unique()
```

```python
new_data.IDENTIFICATION=data.IDENTIFICATION.map({'Not Anemia':0,'Anemia':1})
new_data
```

```python
new_data.to_csv("Anemia.csv")
```
